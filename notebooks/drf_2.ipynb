```python
removed_cols = ['next-next-pos', 'prev-prev-pos', 'next-pos', 'pos', 'prev-pos', 'C1']
cars = cars.drop(removed_cols)
```

```python


# Set the predictors and response;
# set the response as a factor:
cars['lemma'] = cars['lemma'].asfactor()
cars['next-lemma'] = cars['next-lemma'].asfactor()
cars['next-next-lemma'] = cars['next-next-lemma'].asfactor()
cars['next-next-pos'] = cars['next-next-pos'].asfactor()
cars['next-next-shape'] = cars['next-next-shape'].asfactor()
cars['next-pos'] = cars['next-pos'].asfactor()
cars['next-shape'] = cars['next-shape'].asfactor()
cars['next-word'] = cars['next-word'].asfactor()
cars['pos'] = cars['pos'].asfactor()
cars['prev-iob'] = cars['prev-iob'].asfactor()
cars['prev-lemma'] = cars['prev-lemma'].asfactor()
cars['prev-pos'] = cars['prev-pos'].asfactor()
cars['prev-prev-iob'] = cars['prev-prev-iob'].asfactor()
cars['prev-prev-lemma'] = cars['prev-prev-lemma'].asfactor()
cars['prev-prev-pos'] = cars['prev-prev-pos'].asfactor()
cars['prev-prev-shape'] = cars['prev-prev-shape'].asfactor()
cars['prev-prev-word'] = cars['prev-prev-word'].asfactor()
cars['prev-shape'] = cars['prev-shape'].asfactor()
cars['prev-word'] = cars['prev-word'].asfactor()
cars['shape'] = cars['shape'].asfactor()
cars['word'] = cars['word'].asfactor()
cars['tag'] = cars['tag'].asfactor()
#predictors = ["displacement","power","weight","acceleration","year"]
predictors = ['lemma','next-lemma','next-next-lemma', 'next-next-shape', 'next-next-word', 'next-shape', 'next-word', 'prev-iob','prev-lemma', 'prev-prev-iob'
, 'prev-prev-lemma', 'prev-prev-shape', 'prev-prev-word', 'prev-shape', 'prev-word', 'shape', 'word']
#, 'next-next-pos', 'prev-prev-pos', 'next-pos', 'pos', 'prev-pos'

response = "tag"

```

```python
valid
```

```python
from h2o.estimators import H2ORandomForestEstimator
# Split the dataset into a train and valid set:
train, valid = cars.split_frame(ratios=[.8], seed=1234)

# Build and train the model:
cars_drf = H2ORandomForestEstimator(ntrees=10,
                                    max_depth=5,
                                    min_rows=10)
                                    #calibrate_model=True,
                                    #calibration_frame=valid)
                                    #binomial_double_trees=True)
cars_drf.train(x=predictors,
               y=response,
               training_frame=train,
               validation_frame=valid)

# Eval performance:
perf = cars_drf.model_performance()

# Generate predictions on a validation set (if necessary):
pred = cars_drf.predict(valid)
```

```python
y_pred = np.concatenate(pred[0].as_data_frame().iloc[:,0:].values)
```
